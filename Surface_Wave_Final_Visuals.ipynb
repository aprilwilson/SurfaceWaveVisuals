```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
%matplotlib inline
```

```python
##loading in the dataset
dataset=np.genfromtxt('HW131-Data.csv', delimiter=',',skip_header=True)

dataset = dataset.T

xloc=dataset[1] # midpoint loc
layer = dataset[2] # layer number
zloc=dataset[3] # depth
vel=dataset[4] # shear velocity
error=dataset[5] # error
error = error*0.5
```

```python
layer_0 = []
layer_1 = []
layer_2 = []
layer_3 = []

for i in range(len(layer)):
    if layer[i] == 0:
        layer_0.append([xloc[i],zloc[i],vel[i], error[i]])
    elif layer[i] == 1:
        layer_1.append([xloc[i],zloc[i],vel[i], error[i]])
    elif layer[i] == 2:
        layer_2.append([xloc[i],zloc[i],vel[i], error[i]])
    elif layer[i] == 3:
        layer_3.append([xloc[i],zloc[i],vel[i], error[i]])

layer_0 = np.array(layer_0)
layer_1 = np.array(layer_1)
layer_2 = np.array(layer_2)
layer_3 = np.array(layer_3)
                        
layer_0 = layer_0.T
layer_1 = layer_1.T
layer_2 = layer_2.T
layer_3 = layer_3.T
```

```python
##loading in the dataset
dataset2=np.genfromtxt('HW131-Data_depth_ordered.csv', delimiter=',',skip_header=True)

dataset2 = dataset2.T



xloc2=dataset2[1] # midpoint loc
layer2 = dataset2[2] # layer number
zloc2=dataset2[3] # depth
vel2=dataset2[4] # shear velocity
error2=dataset2[5] # error
error2 = error2*0.5
```

```python
layer2_0 = []
layer2_1 = []
layer2_2 = []
layer2_3 = []

for i in range(len(layer2)):
    if layer2[i] == 0:
        layer2_0.append([xloc2[i],zloc2[i],vel2[i], error2[i]])
    elif layer2[i] == 1:
        layer2_1.append([xloc2[i],zloc2[i],vel2[i], error2[i]])
    elif layer2[i] == 2:
        layer2_2.append([xloc2[i],zloc2[i],vel2[i], error2[i]])
    elif layer2[i] == 3:
        layer2_3.append([xloc2[i],zloc2[i],vel2[i], error2[i]])

layer2_0 = np.array(layer2_0)
layer2_1 = np.array(layer2_1)
layer2_2 = np.array(layer2_2)
layer2_3 = np.array(layer2_3)
                        
layer2_0 = layer2_0.T
layer2_1 = layer2_1.T
layer2_2 = layer2_2.T
layer2_3 = layer2_3.T

```

```python

plt.figure(figsize=(12,8))

plt.plot(layer_0[0],layer_0[1], 'k', marker = 'o', label='Reciever Midpoint Locations')
plt.plot(layer_1[0],layer_1[1], 'gray', label='Layer 1')
plt.plot(layer_2[0],layer_2[1], 'purple', label='Layer 2')
plt.plot(layer_3[0],layer_3[1], 'blue', label='Layer 3')

plt.gca().invert_yaxis()

plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.xlabel('Northing (m)', fontsize=22)
plt.ylabel('Depth (m)', fontsize=22)
plt.title('Raw Data Layer Plot', fontsize=28)
plt.legend(prop={'size': 15})
plt.savefig('raw_layer_2.png')
```

```python
#calculate equation for trendline

fit_1 = np.polyfit(layer_1[0], layer_1[1], 15)
poly_1 = np.poly1d(fit_1)

fit_2 = np.polyfit(layer_2[0], layer_2[1], 15)
poly_2 = np.poly1d(fit_2)

fit_3 = np.polyfit(layer_3[0], layer_3[1], 15)
poly_3 = np.poly1d(fit_3)


#add trendline to plot
# plot(layer_0[0], poly_0(layer_0[0]), 'k')

plt.figure(figsize=(12,8))

plt.plot(layer_0[0],layer_0[1], 'k', marker='o', label='Receiver Midpoint Locations')

plt.plot(layer_1[0], poly_1(layer_1[0]), 'gray', label='Layer 1')

plt.plot(layer_2[0], poly_2(layer_2[0]), 'purple', label='Layer 2')

plt.plot(layer_3[0], poly_3(layer_3[0]), 'blue', label='Layer 3')

plt.gca().invert_yaxis()


plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.xlabel('Northing (m)', fontsize=22)
plt.ylabel('Depth (m)', fontsize=22)
plt.title('Least Squares 9th Degree Fit Layer Model', fontsize=28)
plt.legend(prop={'size': 15})
plt.savefig('layer_model_2.png')

print(round(np.average(layer_1[2])))
print(round(np.average(layer_2[2])))
print(round(np.average(layer_3[2])))


```

```python
plt.figure(figsize=(12,8))

plt.plot(layer_0[0],layer_0[1], 'k', marker='o', label='receiver midpoint locations')

plt.plot(layer_1[0], poly_1(layer_1[0]), 'gray', label='608 m/s')

plt.plot(layer_2[0], poly_2(layer_2[0]), 'purple', label='662 m/s')

plt.plot(layer_3[0], poly_3(layer_3[0]), 'blue', label='756 m/s')

plt.gca().invert_yaxis()

plt.fill_between(layer_0[0], layer_0[1], poly_1(layer_1[0]), color='gray')
plt.fill_between(layer_1[0], poly_1(layer_1[0]), poly_2(layer_2[0]), color='purple')
plt.fill_between(layer_2[0], poly_2(layer_2[0]), poly_3(layer_3[0]), color='blue')
#plt.fill_between(layer_3[0], poly_3(layer_3[0]), 120, color='black')

plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.xlabel('Northing (m)', fontsize=22)
plt.ylabel('Depth (m)', fontsize=22)
plt.title('Least Squares 9th Degree Fit Layer Model with Average Shear Wave Velocities', fontsize=22)
plt.legend(prop={'size': 15})
plt.savefig('layer_model_filled_2.png')
```

```python
plt.figure(figsize=(12,8))
plt.plot(layer_1[0], layer_1[2], color='gray', label='Layer 1')
plt.plot(layer_1[0], layer_2[2], color='purple', label = 'Layer 2')
plt.plot(layer_1[0], layer_3[2], color='blue', label = 'Layer 3')
plt.legend(prop={'size': 15})
for i in range(len(layer_1[3])):
    plt.errorbar(layer_1[0,i], layer_1[2,i], yerr=layer_1[3,i], ecolor='lightgrey', capsize=5)
    plt.errorbar(layer_2[0,i], layer_2[2,i], yerr=layer_2[3,i], ecolor='thistle', capsize=5)
    plt.errorbar(layer_3[0,i], layer_3[2,i], yerr=layer_3[3,i], ecolor='lightskyblue', capsize=5)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.title('Velocity Profile of Layers 1,2,3', fontsize=28)
plt.xlabel('Northing (m)', fontsize=22)
plt.ylabel('Velocity (m/s)', fontsize=22)
plt.savefig('velocity_profile_2.png')

plt.show()
```
